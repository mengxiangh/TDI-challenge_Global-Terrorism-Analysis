```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import json
from matplotlib.collections import PatchCollection
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('globalterrorismdb_0919dist.xlsx')
```

```python
df_cut = df[['country_txt']]
```

```python
df_count = df_cut.groupby('country_txt').size().to_frame()
df_count.columns=['count']
```

```python
temp3 = df.groupby(['region_txt','iyear']).size().to_frame()
temp3.reset_index(inplace=True)
```

```python
temp3.columns=['region','year','count']
temp4 = temp3[~temp3['region'].isin(['Australasia & Oceania','Central Asia','East Asia'])]
```

```python
plt.figure(figsize=(18,10))
plt.title('Terrorism attack couns VS Year (excluding Australasia & Oceania,Central Asia,East Asia)',fontsize=18)
plt.xlabel('Year',fontsize=16)
plt.ylabel('Count',fontsize=16)
ax = sns.lineplot(data=temp4,x='year',y='count',hue='region')
plt.show()
```
